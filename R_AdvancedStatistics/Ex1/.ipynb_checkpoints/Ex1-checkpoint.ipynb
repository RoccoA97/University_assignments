**Here vectors are created, holding the lake’s name and all the parameters, then build a dataframe called ```scottish.lakes``` from the vectors**

````r
lake.names      <- c('Ness', 'Lomond', 'Morar', 'Tay', 'Awe', 
                     'Maree', 'Ericht', 'Lochy', 'Rannoch', 'Shiel', 
                     'Katrine', 'Arkaig', 'Shin')
lake.volume     <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
lake.area       <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
lake.length     <- c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8)
lake.max_depth  <- c(230, 190, 310, 150 ,94  ,114 ,156 ,162 ,134 ,128 ,151 ,109, 49)
lake.mean_depth <- c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)
````

````r
scottish.lakes <- data.frame(row.names = lake.names, 
                             lake.volume,
                             lake.area,
                             lake.length,
                             lake.max_depth,
                             lake.mean_depth)
````

````r
scottish.lakes
````

### 1.1. - Highest and lowest volume and area lake <a class="anchor" id="Ex1.1."></a>

**Compute the highest and lowest volumes**

````r
cat("Min volume: ",
    min(scottish.lakes$lake.volume),
    " km^3 (",
    lake.names[which.min(scottish.lakes$lake.volume)], ")\n", sep="")

cat("Max volume: ", 
    max(scottish.lakes$lake.volume), 
    " km^3 (", 
    lake.names[which.max(scottish.lakes$lake.volume)], ")\n", sep="")
````

**Compute the highest and lowest areas.**

````r
cat("Min area: ",
    min(scottish.lakes$lake.area),
    " km^2 (",
    lake.names[which.min(scottish.lakes$lake.area)], ")\n", sep="")

cat("Max area: ", 
    max(scottish.lakes$lake.area), 
    "   km^2 (", 
    lake.names[which.max(scottish.lakes$lake.area)], ")\n", sep="")
````

### 1.2. - Order the frame with respect to the area and determine the two largest area lakes <a class="anchor" id="Ex1.2."></a>

**Order the dataframe in descending order with respect to the area**

````r
scottish.lakes.sorted <- scottish.lakes[order(scottish.lakes$lake.area, decreasing=TRUE),]
scottish.lakes.sorted
````

**Print the two largest area lakes**

````r
cat("1st greatest area: ",
    scottish.lakes.sorted[["lake.area"]][1],
    " km^2 (",
    row.names(scottish.lakes.sorted)[1], ")\n", sep="")

cat("2nd greatest area: ",
    scottish.lakes.sorted[["lake.area"]][2],
    " km^2 (",
    row.names(scottish.lakes.sorted)[2], ")\n", sep="")
````

### 1.3. - Area of Scotland covered by water <a class="anchor" id="Ex1.3."></a>

````r
cat("Sum of all areas: ", 
    sum(scottish.lakes.sorted[["lake.area"]]),
    " km^2 (Scotland)", sep="")
````

**Load the needed libraries**

````r
library(lubridate)
library(readxl)
library(curl)
````

**Get data from web and read**

````r
url    <- "https://www.ecdc.europa.eu/sites/default/files/documents/"
fname  <- "COVID-19-geographic-disbtribution-worldwide-"
date   <-  lubridate::today() - 1
ext    <- ".xlsx"
target <- paste(url, fname, date, ext, sep="")

message("target: ", target)

tmp_file <- tempfile("data", "/tmp", fileext=ext)
tmp      <- curl::curl_download(target, destfile=tmp_file)
````

### 3.2. - Sub-tibble of the last day data <a class="anchor" id="Ex3.2."></a>

**Extract last day data**

````r
covid.lastday <- covid[covid$dateRep==date,]
````

**Filter last day data by cases and deaths, then sort in descending order with respect to the number of cases**

````r
covid.lastday.filtered <- covid.lastday[(covid.lastday$cases>200) | (covid.lastday$deaths>200),]
covid.lastday.filtered <- covid.lastday.filtered[order(covid.lastday.filtered$cases, decreasing = TRUE),]
````

**Put ```covid.lastday.filtered``` (previous cell) into a table and print it**

````r
covid.lastday.filtered.table <- matrix(c(covid.lastday.filtered$cases,
                                         covid.lastday.filtered$deaths),
                                       ncol=2)

colnames(covid.lastday.filtered.table) <- c('cases', 'deaths')
rownames(covid.lastday.filtered.table) <- covid.lastday.filtered$countriesAndTerritories

covid.lastday.filtered.table <- as.table(covid.lastday.filtered.table)
````

### 3.3. - Top 10 countries cases and deaths <a class="anchor" id="Ex3.3."></a>

**Load needed packages**

````r
library(dplyr)
library(ggplot2)
library(repr)
````

**Add cumulative cases and deaths into dataset through a trick using ```ave``` function**

````r
covid$tot_cases  <- ave(covid$cases,  covid$countryterritoryCode, FUN=function(x) rev(cumsum(rev(x))))
covid$tot_deaths <- ave(covid$deaths, covid$countryterritoryCode, FUN=function(x) rev(cumsum(rev(x))))
````

**Extract top 10 countries (in term of cases) last day data**

````r
covid.lastday <- covid[covid$dateRep==date,]

covid.lastday.top10 <- covid.lastday[order(covid.lastday$tot_cases, decreasing = TRUE),]
covid.lastday.top10 <- covid.lastday.top10[1:10,]
````

**Print top 10 countries name**

````r
covid.lastday.top10$countriesAndTerritories
````

**Organize top 10 countries data for next analysis**

````r
covid.top10 <- subset(covid, covid$countriesAndTerritories %in% covid.lastday.top10$countriesAndTerritories)
head(covid.top10)
````

**Normalize data (with respect to the date of the first case) for next analysis**

````r
covid.top10.normalized <- covid.top10[covid.top10$tot_cases>0,]

covid.top10.normalized <- covid.top10.normalized %>%
                          group_by(countriesAndTerritories) %>%
                          mutate(dateRepNormalized = as.numeric(
                              difftime(
                                  dateRep,
                                  last(dateRep), 
                                  units="days")
                              )
                          )
````

**Plot normalized data: some useful tools to simplify formatting**

````r
options(repr.plot.width=12, repr.plot.height=8)
````

````r
min.day <- min(covid.top10.normalized$dateRepNormalized)
max.day <- max(covid.top10.normalized$dateRepNormalized)

min.tot_cases <- min(covid.top10.normalized$tot_cases)
max.tot_cases <- max(covid.top10.normalized$tot_cases)

min.tot_deaths <- min(covid.top10.normalized$tot_deaths)
max.tot_deaths <- max(covid.top10.normalized$tot_deaths)

n.xticks <- 20
n.yticks <- 5
````

**Plot normalized data: number of total cases as a function of days from the beginning of the contagion in a certain territory**

````r
ggplot(covid.top10.normalized, 
       aes(covid.top10.normalized$dateRepNormalized, 
           covid.top10.normalized$tot_cases, 
           group=factor(countriesAndTerritories), 
           color=factor(countriesAndTerritories))
      ) + 
      geom_point(
          shape=1
      ) +
      labs(x="Days",
           y="Number of total cases",
           color="Countries and territories"
      ) + 
      ggtitle(
          "Number of total cases from the beginning of the contagion"
      ) + 
      theme_bw(
      ) + 
      theme(
          plot.title=element_text(hjust=0.5)
      ) +
      scale_x_continuous(
          breaks=round(
              seq(
                  min.day, 
                  max.day, 
                  by=(max.day-min.day) %/% n.xticks
              )
          )
      ) +
      geom_line(
          size=0.1
      )
````

**Plot normalized data: number of total deaths as a function of days from the beginning of the contagion in a certain territory**

````r
ggplot(covid.top10.normalized, 
       aes(covid.top10.normalized$dateRepNormalized, 
           covid.top10.normalized$tot_deaths, 
           group=factor(countriesAndTerritories), 
           color=factor(countriesAndTerritories))
      ) + 
      geom_point(
          shape=1
      ) +
      labs(x="Days",
           y="Number of total deaths",
           color="Countries and territories"
      ) + 
      ggtitle(
          "Number of total deaths from the beginning of the contagion"
      ) + 
      theme_bw(
      ) + 
      theme(
          plot.title=element_text(hjust=0.5)
      ) +
      scale_x_continuous(
          breaks=round(
              seq(
                  min.day, 
                  max.day, 
                  by=(max.day-min.day) %/% n.xticks
              )
          )
      ) +
      geom_line(size=0.1)
````
