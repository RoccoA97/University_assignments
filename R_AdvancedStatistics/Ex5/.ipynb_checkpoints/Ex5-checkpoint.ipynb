**Moreover, let's define some useful functions to solve the following exercises:**

* **```get.x.dmax :``` compute and get the most probable value (mode) of the distribution;**
* **```get.c.int  :``` compute and get the bounds of the XX% credibility interval.**

**```get.x.dmax``` function**

````r
get.x.dmax <- function(x, d) {
    i   <- which.max(d)
    
    if (i == 1 | i==length(x)) {
        return(x[i])
    }
    
    # find most probable value through parabolic approximation
    x1 <- x[i-1]
    x2 <- x[i]
    x3 <- x[i+1]
    y1 <- d[i-1]
    y2 <- d[i]
    y3 <- d[i+1]
    
    den <- (x1 - x2) * (x1 - x3) * (x2 - x3)
    A   <- (x3 * (y2 - y1) + x2 * (y1 - y3) + x1 * (y3 - y2)) / den
    B   <- (x3*x3 * (y1 - y2) + x2*x2 * (y3 - y1) + x1*x1 * (y2 - y3)) / den
    C   <- (x2 * x3 * (x2 - x3) * y1 + x3 * x1 * (x3 - x1) * y2 + x1 * x2 * (x1 - x2) * y3) / den

    x.m <- -B / (2*A);
    
    return(x.m)
}
````

**```get.c.int``` function**

````r
get.c.int <- function(x, d, cl=0.95) {
    delta.x <- x[2] - x[1]
    d       <- d*delta.x
    alpha   <- 1-cl
    
    # find credibility interval bounds through linear approximation
    for (i in 1:(length(d)-1)) {
        if (sum(d[1:i])<alpha/2 & sum(d[1:(i+1)])>alpha/2) {
            m   <- (sum(d[1:(i+1)]) - sum(d[1:i])) / (x[i+1] - x[i])
            q   <- sum(d[1:i])
            x.1 <- (alpha/2 - q) / m + x[i]
        }
        if (sum(d[1:i])<(1-alpha/2) & sum(d[1:(i+1)])>(1-alpha/2)) {
            m   <- (sum(d[1:(i+1)]) - sum(d[1:i])) / (x[i+1] - x[i])
            q   <- sum(d[1:i])
            x.2 <- (1 - alpha/2 - q) / m + x[i]
        }
    }
    
    return(c(x.1,x.2))
}
````

### 1.1. - Assumption for $y$ distribution <a class="anchor" id="Ex1.1."></a>

**I would assume a Binomial distribution for $y$ with the following motivation: if we take one person, considered the informations we have from the text, there is a probability of $p$ that this person has read the journal. So, given $n$ persons, the number of people that have read the journal follows a Binomial distribution with parameters $p$ and $n$.**

### 1.2. - Posterior distribution for $y$ <a class="anchor" id="Ex1.2."></a>

**Let's assume now a uniform prior:**

$$
\mathbb{P}(p|M) \sim U(0,1)
$$

**The posterior distribution for $y$ is given by:**

$$
\mathbb{P}(p|n,y,M)
=
\frac{\mathbb{P}(y|p,n,M) \mathbb{P}(p|M)}{\mathbb{P}(y|n,M)}
=
\frac{1}{Z} p^y (1-p)^{n-y}
$$

**with $Z$ a normalization factor.**

### 2.1. - Plot and comparison of the three prior distributions <a class="anchor" id="Ex2.1."></a>

````r
n_points <- 1001 # number of points in the plots
p        <- seq(0, 1, length.out=n_points)
Dp       <- p[2] - p[1]
````

**Anna's prior**

````r
mu <- 0.2
sd <- 0.08

alpha <-    mu  * (mu*(1-mu)/sd^2 - 1)
beta  <- (1-mu) * (mu*(1-mu)/sd^2 - 1)

ddata.a <- dbeta(p, shape1=alpha, shape2=beta)

options(repr.plot.width=9, repr.plot.height=9)
barplot(height=ddata.a[seq(1,length(p),10)],
        names.arg=p[seq(1,length(p),10)],
        ylim=c(0,1.2*max(ddata.a)),
        xlab='p',
        ylab='density',
        main='Anna\'s prior PDF',
        col='darkorange')
box()
grid()
````

**Benny's prior**

````r
ddata.b <- dunif(p)

options(repr.plot.width=9, repr.plot.height=9)
barplot(height=ddata.b[seq(1,length(p),10)],
        names.arg=p[seq(1,length(p),10)],
        ylim=c(0,1.2*max(ddata.b)),
        xlab='p',
        ylab='density',
        main='Benny\'s prior PDF',
        col='darkorange')
box()
grid()
````

**Chris's prior**

````r
chris.prior <- function(x) {
    res <- ifelse(x>=0 & x<=1,
                  ifelse(x<0.5,
                         ifelse(x<0.3,
                                ifelse(x<0.1,
                                       20*x,
                                       2),
                                5-10*x),
                         0.),
                  NA)
    return(res)
}
````

````r
ddata.c <- chris.prior(p)

options(repr.plot.width=9, repr.plot.height=9)
barplot(height=ddata.c[seq(1,length(p),10)],
        names.arg=p[seq(1,length(p),10)],
        ylim=c(0,1.2*max(ddata.c)),
        xlab='p',
        ylab='density',
        main='Chris\'s prior PDF',
        col='darkorange')
box()
grid()
````

**Comparison of the three priors**

````r
options(repr.plot.width=9, repr.plot.height=9)
plot(p, ddata.a, 
     xaxs='i', yaxs='i', col='navy', 
     type='l', lty=1, lwd=3, 
     ylim=c(0,1.2*max(max(ddata.a), max(ddata.b), max(ddata.c))), 
     xlab="p",
     ylab=expression(paste("P","(p | M)")),
     main='Comparison of prior PDFs')

lines(p, ddata.b, col='firebrick3',lwd=3, lty=1)
lines(p, ddata.c, lwd=3, lty=1, col='seagreen3')

legend("topright",
       inset=0.05, 
       legend=c("Anna","Benny","Chris"),
       col=c("navy", "firebrick3", "seagreen3"), 
       lty=1:1, 
       cex=1,
       box.lty=1,
       lwd=3)
grid()
````

### 3.3. - Sequential analysis of the data <a class="anchor" id="Ex3.3."></a>

**The key idea in this point is the following one:**

* **in the first toss, the priors employed are the same of the previous cells (flat and beta);**
* **in the next tosses, the (normalized) posterior result of the preious toss is used as prior of the new toss.**

**The results for every toss are stored in vectors and then plotted in order to visualize the probability process.**

````r
x.m.f <- NULL
x.m.b <- NULL
x.c.f <- NULL
x.c.b <- NULL
````

````r
n_points <- 1001 # number of points in the plots
p        <- seq(0, 1, length.out=n_points)
Dp       <- p[2] - p[1]
````

**Sequential analysis with ```for``` loop**

````r
for (i in 1:length(tosses)) {
    n.T <- length(tosses[i][tosses[i]=='T'])
    n.H <- length(tosses[i][tosses[i]=='H'])
    
    r <- n.H
    
    if (i!=1) {
        p.prior.f  <- p.post.n.f
        p.prior.b  <- p.post.n.b
    } else {
        p.prior.f  <- dunif(p, min=0, max=1)
        p.prior.b  <- dbeta(p, shape1=15, shape2=15)
    }
    p.post.f   <- dbinom(x=r, size=1, prob=p) * p.prior.f
    p.post.n.f <- p.post.f / (Dp*sum(p.post.f))
    
    p.post.b   <- dbinom(x=r, size=1, prob=p) * p.prior.b
    p.post.n.b <- p.post.b / (Dp*sum(p.post.b))
    
    x.m.f <- c(x.m.f, get.x.dmax(p, p.post.n.f))
    x.m.b <- c(x.m.b, get.x.dmax(p, p.post.n.b))
    x.c.f <- cbind(x.c.f, get.c.int(p, p.post.n.f, cl=0.95))
    x.c.b <- cbind(x.c.b, get.c.int(p, p.post.n.b, cl=0.95))
}
````

**Plot the results for both initial priors: most probable value and 95% credibility interval bounds**

````r
# flat prior
x <- c(1:n)
options(repr.plot.width=12, repr.plot.height=8)
plot(x, x.m.f,
     xaxs='i', yaxs='i', col='firebrick3', 
     type='l', lty=1, lwd = 3, 
     ylim=c(0,1), 
     xlab="#coin tosses",
     ylab=expression(paste("p"["2.5%"], " | ", "p"["m"], " | ", "p"["97.5%"])),
     main='Sequential analysis for initial flat prior')
lines(x, x.c.f[1,], col='navy',lwd=3, lty=1)
lines(x, x.c.f[2,], col='navy',lwd=3, lty=1)
polygon(x=c(x,rev(x)), y=c(x.c.f[1,], rev(x.c.f[2,])), col='seagreen1')
lines(x, x.m.f, col='firebrick3',lwd=3, lty=1)
abline(h=x.m.f.whole, col="firebrick3", lty=2, lwd=1)
abline(h=x.c.f.whole, col="navy",       lty=2, lwd=1)
legend("topright", 
       inset=0.03, 
       legend=c(expression(paste("p"["2.5%"], " | ", "p"["97.5%"], " for sequential analysis")),
                expression(paste("p"["m"], " for sequential analysis")),
                expression(paste("Results with the whole data set"))),
       col=c("navy", "firebrick3", "grey40"), 
       lty=c(1,1,2), 
       cex=1,
       box.lty=1,
       lwd=c(3,3,1))

# beta prior
x <- c(1:n)
options(repr.plot.width=12, repr.plot.height=8)
plot(x, x.m.b,
     xaxs='i', yaxs='i', col='firebrick3', 
     type='l', lty=1, lwd = 3, 
     ylim=c(0,1), 
     xlab="#coin tosses",
     ylab=expression(paste("p"["2.5%"], " | ", "p"["m"], " | ", "p"["97.5%"])),
     main='Sequential analysis for initial beta prior')
lines(x, x.c.b[1,], col='navy',lwd=3, lty=1)
lines(x, x.c.b[2,], col='navy',lwd=3, lty=1)
polygon(x=c(x,rev(x)), y=c(x.c.b[1,], rev(x.c.b[2,])), col='seagreen1')
lines(x, x.m.b, col='firebrick3',lwd=3, lty=1)
abline(h=x.m.b.whole, col="firebrick3", lty=2, lwd=1)
abline(h=x.c.b.whole, col="navy",       lty=2, lwd=1)
legend("topright", 
       inset=0.03, 
       legend=c(expression(paste("p"["2.5%"], " | ", "p"["97.5%"], " for sequential analysis")),
                expression(paste("p"["m"], " for sequential analysis")),
                expression(paste("Results with the whole data set"))),
       col=c("navy", "firebrick3", "grey40"), 
       lty=c(1,1,2), 
       cex=1,
       box.lty=1,
       lwd=c(3,3,1))
````
