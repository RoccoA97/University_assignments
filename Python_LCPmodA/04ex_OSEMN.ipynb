5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

## Exercise 5 <a class="anchor" id="04.5"></a>
* [Index](#04.i)

```python
import json
import csv

data = json.load(open('data/04ex/user_data.json'))

with open('data/04ex/04ex_5.csv', 'w+', newline='') as csvFile:
    writer = csv.writer(csvFile,delimiter=',')
    
    writer.writerow(['ID','JobTitle','EmailAddress','FirstNameLastName','CreditCard','CreditCardType'])
    for d in data:
        if d['CreditCardType']=='American Express':
            writer.writerow([d['ID'],
                             d['JobTitle'],
                             d['EmailAddress'],
                             d['FirstNameLastName'],
                             d['CreditCard'],
                             d['CreditCardType']])
```
